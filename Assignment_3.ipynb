```python
# Importing Pandas and Matplotlib for data viewing and visualisation
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the file as a database
df = pd.read_csv('hotel_bookings(1) (1).csv')
```

```python
# Changing the file's name to Data for easier code
data = pd.read_csv('./hotel_bookings(1) (1).csv')
   
```

```python
# Viewing Data
data
```

```python
# Laying it out in different way to see other statistics
data.describe()
```

```python
# Adding new cancellation category for ease of code
data['did_they_cancel'] = data['is_canceled'] == 1
```

```python
# Calculating the mean lead times of those who cancelled and didn't cancel
average_lead_time_cancelled = df[df['is_canceled'] == 1]['lead_time'].mean()
average_lead_time_not_cancelled = df[df['is_canceled'] == 0]['lead_time'].mean()
```

```python
#Finding the total mean adr for the whole data and naming it
mean_adr = df['adr'].mean()
mean_adr
```

```python
# Finding the average lead time of those who payed above and those who payed below the average adr 
high_adr = df[df['adr'] > mean_adr]
low_adr = df[df['adr'] < mean_adr]

average_lead_time_high_adr = high_adr['lead_time'].mean()
average_lead_time_low_adr = low_adr['lead_time'].mean()
```

```python
# Taking the values of each calculation
print(f"The average lead time for people who paid a daily rate above the average is: {average_lead_time_high_adr:.2f} days")
print(f"The average lead time for people who paid a daily rate below the average is: {average_lead_time_low_adr:.2f} days")
print(f"The average lead time for customers who cancelled their booking is: {average_lead_time_cancelled:.2f} days")
print(f"The average lead time for customers who did not cancel is: {average_lead_time_not_cancelled:.2f} days")
```

```python
# Giving the categories to create a bar chart with the data
categories = ['Above Average ADR', 'Below Average ADR', 'Not Cancelled', 'Cancelled']
average_lead_times = [average_lead_time_high_adr, average_lead_time_low_adr, average_lead_time_not_cancelled, 
                      average_lead_time_cancelled]
```

```python
# Creating the bar chart
plt.figure(figsize=(10, 6))
plt.bar(categories, average_lead_times, color=['green', 'red', 'tab:cyan', 'tab:purple'])

plt.xlabel('Category')
plt.ylabel('Average Lead Time (Days)')
plt.title('Comparison of Average Lead Time for Revenue-Based Categories');

#Naming Bar Chart for Markdown
plt.savefig('rev_bar_chart.png')
```

The question I have chosen is "How does lead time impact revenue streams in the booking industry". I believe that this is important as it can reveal key factors in making profit for businesses, and help them to make decisions moving forward. Lead time is one of the best ways to compare clients as it isn't unique to any singular destination, time, hotel or anything else, allowing for an easy method to analyse operations. For this I required only the lead time, whether they cancelled or not, and the Average Daily Rate (ADR) of the customers. I chose this data specifically as I was looking at revenue, and both ADR and cancellations have major implications on profit and revenue. ![Bar chart](rev_bar_chart.png)The results, shown in this bar chart highlighted that people who paid below the average ADR, generally had a higher lead time than the people who paid above. Similiarly, the average lead time of those who cancelled was significantly higher than those who didn't. These results draw the conclusion that having shorter lead times may generate more financial success for businesses, which they could capitalize on by incentivising customers to book closer to arrival dates, with things like promotions and deals.
