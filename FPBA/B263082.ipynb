```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
#df = pd.read_csv("hotel_bookings(1).csv")
df = pd.read_csv("hotel_bookings.csv")

# Focus on relevant columns
df_clean = df[['lead_time', 'is_canceled']].dropna()
df_clean['lead_time_group'] = pd.cut(df_clean['lead_time'], bins=[0, 30, 60, 90, 180, 365, 600], 
                                     labels=['0-30', '31-60', '61-90', '91-180', '181-365', '366+'])

# Calculate cancellation rate per group
grouped = df_clean.groupby('lead_time_group')['is_canceled'].mean().reset_index()
```

```python
# Supporting Graph: Number of bookings per lead time group
plt.figure(figsize=(10,6))
df_clean['lead_time_group'].value_counts().sort_index().plot(kind='bar', edgecolor='black')
plt.title('Number of Bookings by Lead Time Group')
plt.xlabel('Lead Time (days)')
plt.ylabel('Number of Bookings')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Final visualization
plt.figure(figsize=(10,6))
plt.bar(grouped['lead_time_group'], grouped['is_canceled'], edgecolor='black')
plt.title('Cancellation Rate by Lead Time Group')
plt.xlabel('Lead Time (days)')
plt.ylabel('Cancellation Rate')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Mini-Report

**Business Question:**  
How does the lead time affect the likelihood of cancellation for hotel bookings?

**Relevance:**  
Understanding the relationship between how far in advance a booking is made (lead time) and its likelihood of cancellation is critical for hotel operations. It allows hotels to improve forecasting accuracy, manage inventory and pricing more effectively, and reduce financial risks associated with last-minute cancellations. With this insight, hotels can implement dynamic policies that better align with customer behavior.

**Data Justification:**  
This analysis uses two key variables from the dataset: `lead_time` and `is_canceled`. The `lead_time` represents the number of days between the booking and arrival, while `is_canceled` is a binary indicator of whether the booking was canceled. Grouping bookings by lead time ranges allows us to see cancellation trends across different timeframes.

**Findings and Insights:**  
The visualization shows that bookings with longer lead times tend to have higher cancellation rates. Bookings made 0–30 days before arrival have the lowest cancellation rate, while those made 181+ days in advance have significantly higher rates. This suggests that customers booking far in advance are more likely to change their plans. Hotels might consider stricter cancellation policies or deposits for long-lead bookings to reduce last-minute losses.
