```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt 
```

```python
# Load the dataset 
pd.read_csv("./Data/hotel_bookings(1).csv")
```

```python
df = pd.read_csv("hotel_bookings.csv")
```

```python
df = df[['is_repeated_guest', 'is_canceled']]
```

```python
# Group by repeat guest and calculate cancellation rate 
repeat_guest_cancellation = df.groupby('is_repeated_guest')['is_canceled'].mean() * 100
```

```python
# Clarify names of index values 
repeat_guest_cancellation.index = ['First-time Guests', 'Repeat Guests']
```

```python
# Plot results 
plt.figure(figsize=(6, 5))
repeat_guest_cancellation.plot(kind='bar', color=['#2565be', '#b12796'], edgecolor='black')
plt.xlabel("Guest Type")
plt.ylabel("Cancellation Rate (%)")
plt.title("Cancellation Rate: First-time vs. Repeat Guests")
plt.xticks(rotation=0)
plt.ylim (0, 50)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add text and arrow annotation
plt.annotate("First-time guests more apparent to cancel", 
             xy=(0, repeat_guest_cancellation[0]), xytext=(-0.4, repeat_guest_cancellation[0] + 5), 
             arrowprops=dict(facecolor='Black', arrowstyle='->'), fontsize=9, color='black')


plt.show()

```

# Business Question:
Do repeat guests have a lower cancellation rate than first-time guests and what can be done to reduce cancellation rates?

## Relevance:
This helps us to understand whether repeat guests cancel less frequently and can allow for the development of loyalty programs and offers to reward those who continue to return and incentives for first-timers.

### Data Justification 
'is_repeated_guest' (0 = first-time guest, 1 = repeat guest) and 'is_cancelled' (whether a booking was cancelled or not) were chosen as together they can be directly compared to see whether those who are first time visitors cancel more often or less than repeat visitors. 

#### Findings
The bar chart shows that first-time guests have a higher cancellation rate (around 40%) compared to repeat guests (15%). This suggests that loyalty and familiarity with a hotel reduce the likelihood of cancellations. Hotels can leverage this by offering loyalty programs to encourage return visits. Additionally, implementing stronger cancellation policies, higher deposit requirements, or incentives for first-time guests could reduce cancellations. A forum to address potential concerns, such as unclear booking policies or pricing issues, could also benefit hotels. This would enhance booking reliability, improve customer retention, and ensure a better experience for first-time guests, ultimately reducing cancellations.
