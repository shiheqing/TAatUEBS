```python
import pandas as pd
import matplotlib.pyplot as plt
file = pd.read_csv("hotel_bookings.csv")
file["market_segment"] = file["market_segment"].str.upper().str.strip()
file = file[file["market_segment"] != "UNDEFINED"]
market_cancellation = file.groupby("market_segment").agg(total_bookings=("is_canceled", "size"),cancellations=("is_canceled", "sum"))
print(market_cancellation)
market_cancellation["cancellation_rate"] = market_cancellation["cancellations"] / market_cancellation["total_bookings"]




plt.figure(figsize=(15, 10))
plt.bar(market_cancellation.index, market_cancellation["cancellation_rate"], color='skyblue')
plt.xlabel("Market Segment")
plt.ylabel("Cancellation Rate")
plt.title("Cancellation Rate by Market Segment")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

1.Business Question
Which market segment has the highest cancellation rate?

2.Relevance
Cancellations directly affect hotel revenues and operational efficiency. By identifying high-cancellation market segments, hotel managers gain valuable insights to refine pricing strategies, optimize booking policies, and improve resource allocation. This analysis also helps minimize overbooking risks while enhancing customer retention efforts through targeted approaches tailored to each segment's cancellation patterns.

3.Data Justification
This analysis focuses on two key columns: `market_segment`, which categorizes the source of each booking (such as direct, corporate, or travel agent), and `is_canceled`, which indicates whether a booking was cancelled (1) or not (0). By grouping the data by `market_segment` and calculating the mean of `is_canceled`, we obtain the cancellation rate for each segment

4.Finding and Insights
Data reveals that **group bookings** consistently show the highest cancellation rates compared to other market segments. This trend likely stems from the inherent unpredictability of group travel—events get postponed, plans change, or last-minute adjustments arise.  To counter this, hotels could:  
-Require higher deposits
-Offer flexible packages 
-Factor in cancellation risks
This balances group business with stable revenue.
