```python
import pandas as pd
```

```python
#df=pd.read_excel(r"hotel_bookings(1).xlsx")
df=pd.read_csv(r"hotel_bookings.csv")
```

```python
df
```

```python
grouped = df[['customer_type', 'total_of_special_requests']].groupby('customer_type').mean()
print(grouped)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(grouped.index, grouped['total_of_special_requests'], 
       color='pink', edgecolor='black', alpha=0.7)
ax.set_title('Average Special Requests by Customer Type')
ax.set_xlabel('Customer Type')  
ax.set_ylabel('Average Number of Special Requests')
ax.set_xticks(range(len(grouped.index)))
ax.set_xticklabels(grouped.index, rotation=45)
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(grouped['customer_type'], grouped['total_of_special_requests'], 
       color='pink', edgecolor='black', alpha=0.7)
ax.set_title('Average Special Requests by Customer Type')
ax.set_xlabel('Customer Type')  
ax.set_ylabel('Average Number of Special Requests')
ax.set_xticks(range(len(grouped['customer_type'])))
ax.set_xticklabels(grouped['customer_type'], rotation=45)
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
