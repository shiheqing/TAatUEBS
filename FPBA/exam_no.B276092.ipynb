```python
import pandas as pd
import matplotlib.pyplot as plt

#import dataset
#df = pd.read_csv("hotel_bookings(1).csv")
df = pd.read_csv("hotel_bookings.csv")

# abtract key words
df = df[['hotel', 'adr', 'is_canceled']]

# Group by hotel type for analysis
grouped = df.groupby("hotel")
avg_adr = grouped["adr"].mean()
cancellation_rate = grouped["is_canceled"].mean() * 100

# Create a canvas with two subplots
fig, ax1 = plt.subplots(figsize=(8,6))
bar_width = 0.4
indices = range(len(avg_adr.index))

# Subplot 1: Average Daily Rate (ADR)
bars1 = ax1.bar(indices, avg_adr, bar_width, label='Average Daily Rate (ADR)', color='skyblue')
ax1.set_xlabel("Hotel Type")
ax1.set_ylabel("ADR")
ax1.set_xticks([i + bar_width/2 for i in indices])
ax1.set_xticklabels(avg_adr.index)
ax1.set_title("ADR and Cancellation Rates by Hotel Type")

# Subplot 2: Cancellation Rate
ax2 = ax1.twinx()
bars2 = ax2.bar([i + bar_width for i in indices], cancellation_rate, bar_width, label='Cancellation Rate (%)', color='salmon')
ax2.set_ylabel("Cancellation Rate (%)")

#gather two bars in one graph
bars = bars1 + bars2


# Adjust layout and save
plt.tight_layout()
plt.savefig('hotel_analysis.png', dpi=300, bbox_inches='tight')  # Export as image
plt.show()
```

# Business Question
How do the average daily rate (ADR) and cancellation rates differ between city hotels and resort hotels?  

# Relevance
Differentiating pricing and cancellation behaviors between hotel types is crucial for revenue management and operational planning. Understanding these differences can help hotel managers optimize pricing strategies and improve booking policies.


# Data Justification
- **ADR**: Represents the average daily rate.
- **is_canceled**: A flag indicating whether a booking was canceled.
- **Hotel Type**: The `hotel` column distinguishes between city hotels and resort hotels.

These attributes are ideal for comparing performance and uncovering key insights.
