```python
import pandas as pd
import matplotlib.pyplot as plt

#df=pd.read_csv('hotel_bookings(1).csv')
df=pd.read_csv('hotel_bookings.csv')
df=df.dropna(subset=['reserved_room_type','assigned_room_type'])

different=df['reserved_room_type']!=df['assigned_room_type']
df['room_changed']=different

room_change_table={}
room_change_list=[]

for room_type in df['reserved_room_type'].unique():
    room_data=df[df['reserved_room_type']==room_type]
    
    changed=room_data['room_changed'].sum()
    total=room_data['room_changed'].count()
    not_changed=total-changed
    room_change_table[room_type]={
        'changed':changed,
        'not_changed':not_changed
    }
    percentage_changed=(changed/total)*100
  
    room_change_list.append({
        'room_type':room_type,
        'changed':changed,
        'not_changed':not_changed,
        'percentage_changed':percentage_changed
    })
    
for result in room_change_list:
    room=result['room_type']
    changed=int(result['changed'])
    not_changed=int(result['not_changed'])
    percent=round(result['percentage_changed'],1)
    
    print(f'Room:{room}')
    print(f'Changed:{changed}')
    print(f'Not changed:{not_changed}')
    print('Percentage changed:',percent,'%')
   
x=[]
y=[]

for r in room_change_list:
    room_type=r['room_type']
    percent=round(r['percentage_changed'],1)
    
    x.append(room_type)
    y.append(percent)
    
plt.figure(figsize=(10,6))
plt.plot(x,y,marker='o')
plt.title('Percentage Changed in Different Room Type')
plt.xlabel('Room type')
plt.ylabel('Percent Change')
plt.show()
    
    
    
```

This analysis the frequency of guests of changing different room type from their reserced.By comparing the reserved room type and assigned room type,the percentage change for different type is calcuated.The result shows that for certain type,Room L,the frequency changing is very high,perhaps this is because it is overbooked.This is very important,as this tolds the hotel to manage and keep statistics on rooms and make timely adjustments for unexpected events.For example,during the summer vacaiton,the flow of customers may be particularly large.And unterstanding which rooms are the most popular can help managers to improve the stock control.I select the reserved room type and assigned room  type from the database,as I think this can directly help me to see the change of costomers from teir original plan.Although this doesn't explore the specific reasons for the changes,for example,they booked the wrong room by accident or they think it's too expensive,it plays an important role in the orperation adjustments and further research into guest preference for the room type.
