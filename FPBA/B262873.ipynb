# Hotel Revenue Analysis by Month

This Notebook analyzes hotel bookings to determine which month generates the highest revenue. The goal is to support seasonal marketing decisions using real data from a hotel bookings dataset which has been provided.

```python
## 1. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## 2. Load the Dataset
# Load the CSV dataset into a pandas DataFrame
df = pd.read_csv("hotel_bookings.csv")
```

```python
## 3. Data Preparation
# Set the assumed nightly revenue value
revenue_per_night = 100

# Filter out canceled bookings and create a new DataFrame
df_not_canceled = df[df['is_canceled'] == 0].copy()

# Calculate total nights per booking
df_not_canceled['total_nights'] = df_not_canceled['stays_in_weekend_nights'] + df_not_canceled['stays_in_week_nights']

# Calculate revenue per booking
df_not_canceled['revenue'] = df_not_canceled['total_nights'] * revenue_per_night
```

```python
## 4. Data Analysis
# Group revenue by month
monthly_revenue = df_not_canceled.groupby('arrival_date_month')['revenue'].sum()

# Reorder the months to calendar order
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
monthly_revenue = monthly_revenue.reindex(month_order)
```

```python
## 5. Final Visualization
# Plotting the final bar chart
monthly_revenue.plot(kind='bar', color='skyblue', edgecolor='black', figsize=(10, 6))
plt.title('Monthly Revenue from Hotel Bookings')
plt.xlabel('Month')
plt.ylabel('Revenue (£)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 6. Mini Report

### 1. Business Question  
**Which month generates the highest hotel revenue, and how can this insight inform seasonal marketing strategies?**

### 2. Relevance  
Understanding seasonal trends in hotel revenue allows managers to optimize marketing campaigns, promotions and staff allocation. By identifying high-revenue months managers can use resources effectively, while low-performing months can be targeted for discounts or partnerships to help increase occupancy.

### 3. Data Justification  
To answer this question, we used the following dataset features:  
- `arrival_date_month` to group bookings by calendar month  
- `stays_in_weekend_nights` and `stays_in_week_nights` to calculate the total nights per booking  
- `is_canceled` to filter out bookings that did not result in revenue  

An assumed nightly rate of £100 was used to estimate revenue per booking. By summing this revenue for all valid bookings per month, we produced a monthly revenue breakdown.

### 4. Findings and Insights  
Our analysis shows that **August** generates the highest revenue, followed closely by **July**. This suggests peak business occurs in late summer, likely due to holidays and travel demand. Hotels could employ more staff, raise prices slightly or focus marketing efforts around these months to maximize profit. On the other hand, lower-revenue months may benefit from targeted promotions to increase bookings.
