```python
# Import the Pandas and Matplotlib libraries into the Python environment
import pandas as pd
import matplotlib.pyplot as plt

# Load the csv file
bookings = pd.read_csv('hotel_bookings.csv')
```

```python
bookings.head(10)
```

```python
bookings.shape
```

```python
bookings.describe()
```

```python
bookings['meal'].unique()
```

```python
# Creating a variable to store a value for how many times each catering option('meal') has been selected
meal_count = (bookings['meal']).value_counts()
```

```python
meal_count.index
```

```python
# Creating a loop to find the total for each catering option
for meal_bookings in bookings['meal'].unique():
    mask = (bookings['meal'] == meal_bookings)
    sum_meals = bookings[mask]['meal'].value_counts()
    print(f'The total number of bookings for is {sum_meals}');
```

```python
# Preparing the data for the pie chart
data_for_chart2 = bookings.groupby('meal').size()
```

```python
data_for_chart2.index;
```

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 0.9, 0.8]) # main axes
axes2 = fig.add_axes([0.4, 0.3, 0.4, 0.4]) # inset axes

# Larger figure Axes 1
axes1.bar(meal_count.index, meal_count,
         color = ['red', 'orange', 'yellow', 'green', 'blue'],
         alpha = 0.7, width = 0.5, align = 'center',
         edgecolor = 'black', linewidth = 0.5,) # Plotting a bar chart on Axes 1 to display the data of the meal_count variable

# Setting the X-axis labels to be an equal distance from each other and to lie at a 45 degree angle
axes1.set_xticks(range(len(meal_count)), meal_count.index,
                 rotation=45)
# Adding a title and labels to the axes of the bar chart
axes1.set_xlabel('Type of Meal')
axes1.set_ylabel('Number of Bookings')
axes1.set_title('Number of bookings including each type of catering')

# Adding annotations to the graph to explain the data
axes1.annotate('Bed and Breakfast\n is the most\n popular option\n with 92313 bookings\n including this option.', xy=('BB', 92313), xytext=('HB', 79000),
             fontsize=7, style='normal', color='black', ha='center',
            arrowprops=dict(facecolor= 'black',));

axes1.annotate('Full board\n is the least\n popular option\n with only 798\n bookings including\n this option.', xy=('FB', 798), xytext=('FB', 10000),
             fontsize=7, style='normal', color='black', ha='center',
            arrowprops=dict(facecolor= 'black',));

axes1.annotate('1169 bookings\n have left their\n catering plan\n as Undefined.', xy=('Undefined', 1169), xytext=('Undefined', 12000),
             fontsize=7, style='normal', color='black', ha='center',
            arrowprops=dict(facecolor= 'black',));

axes1.annotate('Half board is\n the second most\n popular option\n with 14464 bookings\n including this catering. ', xy=('HB', 14464), xytext=('HB', 40000),
             fontsize=7, style='normal', color='black', ha='center',
            arrowprops=dict(facecolor= 'black',));

axes1.annotate('Self-Catering is\n the third most\n popular option\n with 10650 bookings\n including this\n catering option. ', xy=('SC', 10650), xytext=('SC', 30000),
             fontsize=7, style='normal', color='black', ha='center',
            arrowprops=dict(facecolor= 'black',));

axes1.annotate('Bed and Breakfast\n is the catering\n option selected\n the most with\n 77.3% of bookings\n having this\n option selected.', xy=(3.3, 75000), xytext=('FB', 77000),
             fontsize=7, style='normal', color='black', ha='center',
            arrowprops=dict(facecolor= 'black',));

axes1.annotate('The other 4\n catering options\n were selected\n on 22.7%\n of bookings.', xy=(3.4, 50000), xytext=('FB', 35000),
             fontsize=7, style='normal', color='black', ha='center',
            arrowprops=dict(facecolor= 'black',));

# Insert Figure Axes 2
axes2.pie(data_for_chart2, labels=data_for_chart2.index, autopct='%1.1f%%', radius=1); # Plotting a Pie Chart to display the data for the selection percentages of each catering option
```
