```python
# Import the required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('hotel_bookings.csv')

# Count bookings per month and order them chronologically
monthly_bookings = df["arrival_date_month"].value_counts().reindex(["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"])

# Create a bar chart showing the monthly hotel demand
plt.figure(figsize=(10, 5))
plt.bar(monthly_bookings.index, monthly_bookings.values, color="royalblue")
plt.xlabel("Month")
plt.ylabel("Number of Bookings")
plt.title("Hotel Demand Trends Over Month")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Hotel Demand Trends Over Months

**Business Question:**  
How does hotel demand vary over the months of the year? In particular, what seasonal patterns can be observed in the total number of bookings?

**Relevance:**  
Understanding monthly or seasonal demand is crucial for hotel management. By using this data for forecasting, it enables decision-makers to adjust staffing, supplies, and other resources based on anticipated demand and tailor marketing promotions during peak and off-peak periods. Also rates can be adjusted to maximize revenue during peak seasons and attract customers during off-peak seasons.

**Data Justification:**  
This analysis utilized the “arrival_date_month” column to identify when bookings are scheduled to occur. By counting the number of bookings (i.e., demand) per month, we can assess how different times of the year affect hotel utilization and create a robust picture of seasonal trends.

**Findings and Insights:**  
The line chart displays the monthly trend in hotel bookings. The visualization shows that in the summer months, particularly July and August, there is a peak in demand. It might be due to favorable weather conditions and holiday travel. In contrast, the months with the lowest booking volumes are January, November and December suggesting a quieter period for hotel business. This could be attributed to colder weather and fewer holidays. These observations confirm expected seasonality in the hospitality industry.
